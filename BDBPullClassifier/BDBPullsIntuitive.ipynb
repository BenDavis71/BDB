```python
def calculate_adjacent_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<=1) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(position+'S')>2) &
                (pl.col(one_right+'Pull')=='R') &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_right+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_right+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('AR')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(position+'S')>2) &
                (pl.col(one_left+'Pull')=='L') &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_left+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_left+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('AL')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```

```python
def calculate_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    df = df.with_columns([
        pl.lit(None).alias(column_name)
    ])
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(one_right+'RelativeX')<1) &
                (pl.col(position+'AdjustedX')>pl.col(one_right+'AdjustedX')-.5) &
                (pl.col(position+'S')>2)
            ).then('R')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(one_left+'RelativeX')>-1) &
                (pl.col(position+'AdjustedX')<pl.col(one_left+'AdjustedX')+.5) &
                (pl.col(position+'S')>2)
            ).then('L')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```
