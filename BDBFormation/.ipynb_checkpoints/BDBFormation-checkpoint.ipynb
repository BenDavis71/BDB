```python
tracking = pl.read_csv('nfl-big-data-bowl-2024/tracking_week*.csv',infer_schema_length=10000)
tracking = tracking.with_columns(
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str)).alias('uniquePlayId'),
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str) + '-'
     + pl.col('nflId').cast(str)).alias('uniquePlayerId'),
)
# tracking=tracking.filter(pl.col('club')!='football')
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('playDirection')=='right').then(53.3-pl.col('y')).otherwise(pl.col('y')).alias('adjustedX'),
    pl.when(pl.col('playDirection')=='right').then(pl.col('x')).otherwise(120-pl.col('x')).alias('adjustedY'),
#     pl.when(pl.col('playDirection')=='right').then(pl.col('dir').cast(pl.Float64)).otherwise(180-pl.col('dir').cast(pl.Float64)).alias('adjustedDir'),
#     pl.when(pl.col('playDirection')=='right').then(pl.col('o').cast(pl.Float64)).otherwise(180-pl.col('o').cast(pl.Float64)).alias('adjustedO'),
])

# tracking=tracking.with_columns([
#     pl.when(pl.col('adjustedDir')<0).then(360+pl.col('adjustedDir')).otherwise(pl.col('adjustedDir')).alias('adjustedDir'),
#     pl.when(pl.col('adjustedO')<0).then(360+pl.col('adjustedO')).otherwise(pl.col('adjustedO')).alias('adjustedO'),
# ])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('event')=='ball_snap').then(pl.col('frameId')).otherwise(-1).alias('startingFrameId'),
])
tracking=tracking.with_columns([
    pl.col('startingFrameId').max().over(pl.col('uniquePlayId')),
])
tracking=tracking.with_columns([
    (pl.col('frameId') - pl.col('startingFrameId')).alias('framesSinceSnap'),
])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedX')).otherwise(-1000).alias('startingX'),
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedY')).otherwise(-1000).alias('startingY')
])
tracking=tracking.with_columns([
    pl.col('startingX').max().over(pl.col('uniquePlayerId')),
    pl.col('startingY').max().over(pl.col('uniquePlayerId'))
])
tracking=tracking.with_columns([
    (pl.col('adjustedX') - pl.col('startingX')).alias('relativeX'),
    (pl.col('adjustedY') - pl.col('startingY')).alias('relativeY'),
])
# tracking=tracking.with_columns([
#     pl.when(pl.col('club')=='football').then(pl.col('startingX')).otherwise(-1000).alias('footballStartingX')
# ])
# tracking=tracking.with_columns([
#     pl.col('footballStartingX').max().over(pl.col('uniquePlayId'))
# ])
tracking
```

```python
# in case of only 5 linemen, then 1 left of center is lg, 1 right is rg, 2 left is lt if applicable, etc
linemen = linemen.with_columns(
    pl.when(pl.col('position').is_in(['C','G','T', 'c'])).then(1).otherwise(0)
    .sum().over(pl.col('uniquePlayId')).alias('linemenCount')
)

standard = linemen.filter(pl.col('linemenCount')==5).filter(pl.col('position').is_in(['C','G','T', 'c']))
standard = standard.with_columns([
    pl.col('startingX').rank().over('uniquePlayId').alias('linemanOrder')
])

standard = standard.with_columns([
    pl.when(
        (pl.col('linemanOrder')==3) & (pl.col('position')=='c')
    )
    .then(1)
    .otherwise(0)
    .sum()
    .over(pl.col('uniquePlayId'))
    .alias('balanced')
])


standard.filter(pl.col('balanced')==0).join(plays,on=['gameId','playId']).select('gameId','playId','possessionTeam','defensiveTeam','week',
                                           'quarter','gameClock','position','jerseyNumber','linemanOrder','displayName',
                                           'ballX','ballY','startingX','startingY','disFromBall').filter(pl.col('position')=='c')
```

```python
balanced = standard.filter(pl.col('balanced')==1)
balanced = balanced.with_columns([
    pl.col('startingX').rank().over('uniquePlayId').map_dict(linemen_positions).alias('onFieldPosition'),
    pl.lit('balanced').alias('lineBalance')
])
balanced
```

```python
x_box_window = 3
y_box_window = 2.7

df = df.with_columns([
    pl.col('lineBalance').max().over('uniquePlayId'),
    (pl.when(pl.col('onFieldPosition')=='rt').then(pl.col('xFromBall')).otherwise(0))
        .max().over('uniquePlayId').alias('rightTackleBoxX'),
    (pl.when(pl.col('onFieldPosition')=='lt').then(pl.col('xFromBall')).otherwise(0))
        .min().over('uniquePlayId').alias('leftTackleBoxX'),
    (pl.when(pl.col('onFieldPosition')=='c').then(pl.col('yFromBall')-y_box_window).otherwise(0))
        .min().over('uniquePlayId').alias('boxY'),
])  

df = df.with_columns([
    (pl.col('rightTackleBoxX')+x_box_window).alias('rightBoxX'),
    (pl.col('leftTackleBoxX')-x_box_window).alias('leftBoxX')
])

#Remove this to start incorporating 6OL and unbalanced
df = df.filter(pl.col('lineBalance')=='balanced')

df
```

```python
df = df.with_columns([
    pl.when(pl.col('onFieldPosition') == 'ry')
    .then(pl.col('xFromBall'))
    .otherwise(1000000)
    .rank().over('uniquePlayId')
    .cast(int)
    .cast(str)
    .alias('ryRank'),
    pl.when(pl.col('onFieldPosition') == 'ly')
    .then(-pl.col('xFromBall'))
    .otherwise(1000000)
    .rank().over('uniquePlayId')
    .cast(int)
    .cast(str)
    .alias('lyRank'),
])
 
df = df.with_columns([
    pl.when(pl.col('onFieldPosition') == 'ry')
    .then(pl.col('onFieldPosition') + pl.col('ryRank'))
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition'),
])
df = df.with_columns([
    pl.when(pl.col('onFieldPosition') == 'ly')
    .then(pl.col('onFieldPosition') + pl.col('lyRank'))
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition'),
])

df.filter(pl.col('onFieldPosition')=='ly3').join(
    plays.select(['gameId','playId','possessionTeam','defensiveTeam'
                  ,'week','quarter','gameClock']),
    on=['gameId','playId']
).sort(pl.col('gameId'), 'playId')
```

```python
df = df.with_columns([
    pl.when(pl.col('onFieldPosition') == 'rt')
    .then(pl.col('xFromBall'))
    .otherwise(0)
    .max().over(pl.col('uniquePlayId'))
    .alias('rtX'),
    pl.when(pl.col('onFieldPosition') == 'lt')
    .then(pl.col('xFromBall'))
    .otherwise(0)
    .min().over(pl.col('uniquePlayId'))
    .alias('ltX'),
])

df = df.with_columns([
    pl.when(
        (pl.col('xFromBall') > pl.col('rtX')) &
        (pl.col('club') == pl.col('possessionTeam'))
    )
    .then(pl.col('yFromBall'))
    .otherwise(None)
    .max().over(pl.col('uniquePlayId'))
    .alias('rOnBallY'),
    
    
    pl.when(
        (pl.col('xFromBall') < pl.col('ltX')) &
        (pl.col('club') == pl.col('possessionTeam'))
    )
    .then(pl.col('yFromBall'))
    .otherwise(None)
    .max().over(pl.col('uniquePlayId'))
    .alias('lOnBallY'),
])

df = df.with_columns([ 
    pl.when(
        (pl.col('xFromBall') > pl.col('rtX')) &
        (pl.col('club') == pl.col('possessionTeam')) &
        (pl.col('yFromBall') == pl.col('rOnBallY'))
    )
    .then(1)
    .when(
        (pl.col('xFromBall') < pl.col('ltX')) &
        (pl.col('club') == pl.col('possessionTeam')) &
        (pl.col('yFromBall') == pl.col('lOnBallY'))
    )
    .then(1)
    .otherwise(0)
    .alias('onBall')
])

df.select(pl.col('onBall').sum().over('uniquePlayId').max())
# df
```

```python
# get 1st TEs outside of tackles
defense = defense.with_columns([
    pl.when(pl.col('ly1StartingX') + .5 < pl.col('ltStartingX'))
    .then(pl.col('ly1StartingX'))
    .when(pl.col('ly2StartingX') + .5 < pl.col('ltStartingX'))
    .then(pl.col('ly2StartingX'))
    .when(pl.col('ly3StartingX') + .5 < pl.col('ltStartingX'))
    .then(pl.col('ly3StartingX'))
    .otherwise(None)
    .alias('lyStartingX'),
    pl.when(pl.col('ry1StartingX') - .5 > pl.col('rtStartingX'))
    .then(pl.col('ry1StartingX'))
    .when(pl.col('ry2StartingX') - .5 > pl.col('rtStartingX'))
    .then(pl.col('ry2StartingX'))
    .when(pl.col('ry3StartingX') - .5 > pl.col('rtStartingX'))
    .then(pl.col('ry3StartingX'))
    .otherwise(None)
    .alias('ryStartingX'),
])
```

```python
#TODO logic to only check TE if he's not within tackle box
#How would this handle nulls?
offensiveSurface = ['ly'] + linemen + ['ry']
distanceColNames=[]
for player in offensiveSurface:
    distanceColNames.append(player+'Distance') 
    defense = defense.with_columns([
        (pl.col('startingX') - pl.col(player + 'StartingX')).alias(distanceColNames[-1])
    ])
    
defense = defense.with_columns(pl.min([abs(pl.col(colName)) for colName in distanceColNames])
                            .alias('closestDistance'))

defense = defense.with_columns([
    pl.lit(None).alias('closestPlayer')
])

for distanceColName in distanceColNames:
    defense = defense.with_columns([
        pl.when(pl.col('closestDistance')==abs(pl.col(distanceColName)))
        .then(distanceColName[:-8]) # get position name from colName minus 'distance'
        .otherwise(pl.col('closestPlayer'))
        .alias('closestPlayer')
    ])
defense
```

```python
headUpBuffer=.225
wideNineBuffer=2
techniqueDict = {
                 'ly right':'7', 'ly left':'9','ly headUp':'6',
                 'lt right':'4i', 'lt left':'5','lt headUp':'4',
                 'lg right':'2i', 'lg left':'3','lg headUp':'2',
                 'c right':'1', 'c headUp':'0', 'c left':'1',
                 'rg right':'3', 'rg left':'2i','rg headUp':'2',
                 'rt right':'5', 'rt left':'4i','rt headUp':'4',
                 'ry right':'9', 'ry left':'7','ry headUp':'6',
                }

#TODO backers
#TODO only allow one of each technique on each side of the defense
#TODO test the buffer
defense = defense.with_columns([
    pl.when(pl.col('distance')>headUpBuffer)
    .then('right')
    .when(pl.col('distance')<-headUpBuffer)
    .then('left')
    .otherwise('headUp')
    .alias('alignment')
])

defense = defense.with_columns([
        (pl.col('closestPlayer') + pl.lit(' ') + pl.col('alignment'))
        .map_dict(techniqueDict)
        .alias('technique')
    ])

# special cases with techniques
defense = defense.with_columns([
    
    #handle wide nines
    pl.when(
        (abs(pl.col('distance'))>wideNineBuffer) & (pl.col('technique')=='5')
    )
    .then('w9')
    
    #differentiate which gap the 9 tech is in
    .when(
        (pl.col('lyStartingX')==pl.col('ly1StartingX')) 
        & (pl.col('startingX')<(pl.col('ly2StartingX')-headUpBuffer))
        & (pl.col('technique')=='9')
    )
    .then('9e') # showing that he's over the e gap
    
    #same for right side
    .when(
        (pl.col('ryStartingX')==pl.col('ry1StartingX')) 
        & (pl.col('startingX')>(pl.col('ry2StartingX')+headUpBuffer))
        & (pl.col('technique')=='9')
    )
    .then('9e') # showing that he's over the e gap
    .otherwise(pl.col('technique'))
    .alias('technique')
])

defense
```

```python
def calculate_te_pulls(df, position):
    
    column_name = position +'Pull'
    
    df = df.with_columns([
        pl.lit(None).alias(column_name)
    ])
    
    if position.startswith('l'):
          
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1.5) &
                (pl.col(position+'AdjustedY')<pl.col('ltAdjustedY')) &
                (pl.col(position+'RelativeX')>1) &
                #(pl.col(one_right+'RelativeX')<1) &
                (pl.col(position+'AdjustedX')>pl.col('ltAdjustedX')-1) #&
                #(pl.col(position+'S')>2)
            ).then('R')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    elif position.startswith('r'):
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1.5) &
                (pl.col(position+'AdjustedY')<pl.col('rtAdjustedY')) &
                (pl.col(position+'RelativeX')<-1) &
                #(pl.col(one_left+'RelativeX')>-1) &
                (pl.col(position+'AdjustedX')<pl.col('rtAdjustedX')+1) #&
                #(pl.col(position+'S')>2)
            ).then('L')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    
    # counter bluff check
    df = df.with_columns([
        pl.when(
            (pl.col(column_name)=='R') &
            (pl.col(position+'BluffCheckX')<pl.col(position+'AdjustedX')-4)
        ).then('BL') #Bluff Left
        .when(
            (pl.col(column_name)=='L') &
            (pl.col(position+'BluffCheckX')>pl.col(position+'AdjustedX')+4)
        ).then('BR') #Bluff Right
        .otherwise(pl.col(column_name)).alias(column_name)
    ])
    
    
    return df
```

```python
def calculate_adjacent_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<=1) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(position+'S')>2) &
                (pl.col(one_right+'Pull')=='R') &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_right+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_right+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('AR')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(position+'S')>2) &
                (pl.col(one_left+'Pull')=='L') &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_left+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_left+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('AL')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```

```python
def calculate_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    df = df.with_columns([
        pl.lit(None).alias(column_name)
    ])
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(one_right+'RelativeX')<1) &
                (pl.col(position+'AdjustedX')>pl.col(one_right+'AdjustedX')-.5) &
                (pl.col(position+'S')>2)
            ).then('R')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(one_left+'RelativeX')>-1) &
                (pl.col(position+'AdjustedX')<pl.col(one_left+'AdjustedX')+.5) &
                (pl.col(position+'S')>2)
            ).then('L')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```
