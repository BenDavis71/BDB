```python
tracking = pl.read_csv('../nfl-big-data-bowl-2024/tracking_week*.csv',infer_schema_length=10000)
tracking = tracking.with_columns(
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str)).alias('uniquePlayId'),
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str) + '-'
     + pl.col('nflId').cast(str)).alias('uniquePlayerId'),
)
# tracking=tracking.filter(pl.col('club')!='football')
tracking
```

```python
# normalize xy coordinates with respect to playDirection
# transpose x and y so that they're more understandable from a player / coach perspective
tracking=tracking.with_columns([
    pl.when(pl.col('playDirection')=='right').then(53.3-pl.col('y')).otherwise(pl.col('y')).alias('adjustedX'),
    pl.when(pl.col('playDirection')=='right').then(pl.col('x')).otherwise(120-pl.col('x')).alias('adjustedY')
])

# normalize orientation 'o'
# convert 'NA' to 0
replacement_values = {'NA': '0'}
tracking = tracking.with_columns(
    pl.col('o').apply(lambda x: replacement_values.get(x, x)),
)

tracking=tracking.with_columns([
    pl.when(pl.col('playDirection')=='right').then(pl.col('o').cast(pl.Float64)).otherwise((180+pl.col('o').cast(pl.Float64))%360).alias('firstAdjustedO'),
])

tracking=tracking.with_columns([
    pl.when(pl.col('firstAdjustedO') <= 180).then(180-pl.col('firstAdjustedO')).otherwise(540-pl.col('firstAdjustedO')).alias('adjustedO')
])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('event').is_in(['ball_snap','autoevent_ballsnap'])).then(pl.col('frameId')).otherwise(-1)
    .alias('startingFrameId'),
])
tracking=tracking.with_columns([
    pl.col('startingFrameId').max().over(pl.col('uniquePlayId')),
])
tracking=tracking.with_columns([
    (pl.col('frameId') - pl.col('startingFrameId')).alias('framesSinceSnap'),
])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedX')).otherwise(-1000).alias('startingX'),
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedY')).otherwise(-1000).alias('startingY')
])
tracking=tracking.with_columns([
    pl.col('startingX').max().over(pl.col('uniquePlayerId')),
    pl.col('startingY').max().over(pl.col('uniquePlayerId'))
])
tracking=tracking.with_columns([
    (pl.col('adjustedX') - pl.col('startingX')).alias('relativeX'),
    (pl.col('adjustedY') - pl.col('startingY')).alias('relativeY'),
])
tracking
```

```python
df = df.with_columns([
    pl.coalesce([pl.col('onFieldPosition'),pl.lit('na')])
    .alias('onFieldPosition')
])
```

```python
quickMotionSpeedThreshold = 1.75
df = df.with_columns([
    pl.when(pl.col('s')>=quickMotionSpeedThreshold).then(1).otherwise(0)
    .alias('quickMotion')
])
```

```python
xBoxWindow = 2.7
yBoxWindow = 3

df = df.with_columns([
    pl.col('lineBalance').max().over('uniquePlayId'),
    (pl.when(pl.col('onFieldPosition')=='rt').then(pl.col('xFromBall')).otherwise(0))
        .max().over('uniquePlayId').alias('rightTackleBoxX'),
    (pl.when(pl.col('onFieldPosition')=='lt').then(pl.col('xFromBall')).otherwise(0))
        .min().over('uniquePlayId').alias('leftTackleBoxX'),
    (pl.when(pl.col('onFieldPosition')=='c').then(pl.col('xFromBall')).otherwise(10000))
        .min().over('uniquePlayId').alias('cX'),
    (pl.when(pl.col('onFieldPosition')=='c').then(pl.col('yFromBall')).otherwise(10000))
        .min().over('uniquePlayId').alias('cY'),
    (pl.when(pl.col('onFieldPosition')=='c').then(pl.col('club')).otherwise(''))
        .max().over('uniquePlayId').alias('offensiveClub'),
])  

df = df.with_columns([
    (pl.col('rightTackleBoxX')+xBoxWindow).alias('rightBoxX'),
    (pl.col('leftTackleBoxX')-xBoxWindow).alias('leftBoxX'),
    (pl.col('cY')-yBoxWindow).alias('boxY')
])

#Remove this to start incorporating 6OL and unbalanced
df = df.filter(pl.col('lineBalance')=='balanced')

df
```

```python
df = df.with_columns(
    pl.when(
        (pl.col('onFieldPosition').is_in(['y','hb']))
    )
    .then('')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition'),  
)
```

```python
df = df.with_columns(
    pl.when( # any non-qb and non-heavy skill between the tackles with enough depth is hb, including jet WRs
        (~(pl.col('position').is_in(['QB','TE','FB'])))
        & pl.col('xFromBall').is_between(pl.col('leftTackleBoxX'),pl.col('rightTackleBoxX'))
        & (pl.col('yFromBall') < (pl.col('boxY')-1))
    )
    .then('hb')
    .when( # expand the boundaries if they're in quick motion so we can include jet WRs
        (~(pl.col('position').is_in(['QB','TE','FB'])))
        & pl.col('xFromBall').is_between(pl.col('leftTackleBoxX')-6,pl.col('rightTackleBoxX')+6)
        & (pl.col('yFromBall') < (pl.col('boxY')+1))
        & (pl.col('quickMotion')==1)
    )
    .then('hb')
    .when( # backs anywhere on the field, but deep (might miss on some Malzahn-like gadget plays)
        (pl.col('position')=='RB')
        & (pl.col('yFromBall') < (pl.col('boxY')-1))
    )
    .then('hb')
    .when( # backs between the tackles but close to the line (like in pass pro)
        (pl.col('position')=='RB')
        & pl.col('xFromBall').is_between(pl.col('leftTackleBoxX'),pl.col('rightTackleBoxX'))
        & (pl.col('yFromBall') >= pl.col('boxY')-1)
        & (pl.col('quickMotion')==0)
    )
    .then('hb')
    .when( # Ben Skowronek / the Fridge exception
        (~(pl.col('position').is_in(['QB','TE','FB'])))
        & pl.col('xFromBall').is_between(pl.col('leftTackleBoxX'),pl.col('rightTackleBoxX'))
        & (pl.col('yFromBall') < pl.col('boxY')+.5)
        & (pl.col('quickMotion')==0)
    )
    .then('hb')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition')
)
```

```python
# incrementally expand the area to search for more TEs past the current end of the box / offensive surface
for _ in range(4):
    df = df.with_columns(
        pl.when(
            (pl.col('position').is_in(['TE','FB']))
            & pl.col('xFromBall').is_between(pl.col('leftBoxX'),pl.col('rightBoxX'))
            & (pl.col('yFromBall') >= pl.col('boxY'))
        )
        .then('y')
        .otherwise(pl.col('onFieldPosition'))
        .alias('onFieldPosition'),  
    )

    df = df.with_columns(
        (pl.when(pl.col('onFieldPosition')=='y').then(pl.col('xFromBall')+xBoxWindow)
         .otherwise(pl.col('rightBoxX')))
            .max().over('uniquePlayId').alias('rightBoxX'), 
        (pl.when(pl.col('onFieldPosition')=='y').then(pl.col('xFromBall')-xBoxWindow)
         .otherwise(pl.col('leftBoxX')))
            .min().over('uniquePlayId').alias('leftBoxX'),
    )
```

```python
df = df.with_columns(
    pl.when(
        (pl.col('position').is_in(['TE','FB']))
        & pl.col('xFromBall').is_between(pl.col('leftBoxX'),pl.col('rightBoxX'))
        & (~(pl.col('onFieldPosition') == 'y'))
    )
    .then('f')
    .when(
        (pl.col('xFromBall') >= 0)
        & (pl.col('onFieldPosition') == 'y')
    )
    .then('ry')
    .when(
        (pl.col('xFromBall') < 0)
        & (pl.col('onFieldPosition') == 'y')
    )
    .then('ly')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition'),  
)
```

```python
df = df.with_columns(
    pl.when(
        (~(pl.col('xFromBall').is_between(pl.col('leftTackleBoxX'),
                                          pl.col('rightTackleBoxX'))))
        & (~(pl.col('onFieldPosition').is_in(['ry','ly','hb','f'])))
    )
    .then('wr')
    .when(
        (~(pl.col('position').is_in(['QB','TE','FB','RB','T','G','C'])))
        & (~(pl.col('onFieldPosition').is_in(['ry','ly','hb','f'])))
        & pl.col('xFromBall').is_between(pl.col('leftTackleBoxX'),pl.col('rightTackleBoxX'))
        & (pl.col('yFromBall') >= pl.col('boxY')-1)
    )
    .then('wr')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition'),  
)
```

```python
behindCenterBuffer = 1
df = df.with_columns([
    pl.when(
        (pl.col('xFromBall') > (pl.col('cX')-behindCenterBuffer))
        & (pl.col('xFromBall') < (pl.col('cX')+behindCenterBuffer))
        & (pl.col('club') == pl.col('offensiveClub'))
        & (~(pl.col('onFieldPosition').is_in(['c','lg','rg'])))
    )
    .then(1)
    .otherwise(0)
    .alias('behindCenter')
])

df = df.with_columns([
    (pl.when(
        (pl.col('quickMotion') == 0)
        & (pl.col('behindCenter') == 1)
    )
    .then(pl.col('yFromBall'))
    .otherwise(-100))
    .max()
    .over('uniquePlayId')
    .alias('closestYBehindCenter')
])

df = df.with_columns([
    pl.when(
        (pl.col('yFromBall') == pl.col('closestYBehindCenter'))
        & (pl.col('behindCenter') == 1)
    )
    .then('q')
    .when((pl.col('onFieldPosition')=='na') & (pl.col('position')=='QB'))
    .then('hb') # if it's wildcat and the QB ain't a WR, he's a back
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition')
])

df
```

```python
# remove defensive players from onFieldPosition classifications and quickMotion
df = df.with_columns([
    pl.when(pl.col('club')==pl.col('offensiveClub'))
    .then(pl.col('onFieldPosition'))
    .otherwise(None),
    
    pl.when(pl.col('club')==pl.col('offensiveClub'))
    .then(pl.col('quickMotion'))
    .otherwise(None),
])
```

```python
shotgunThreshold = -2.5

df = df.with_columns([
    pl.when(
        (pl.col('yFromBall') >= shotgunThreshold)
        & (pl.col('onFieldPosition') == 'q')
    )
    .then(1)
    .otherwise(0)
    .max().over('uniquePlayId')
    .alias('underCenter')
])
```

```python
# define qb alignment and wildcat / skill player sneak situations
df = df.with_columns([
    pl.when(pl.col('underCenter')==1)
    .then('underCenter')
    .when(
        (pl.col('behindCenter')==1)
        & (pl.col('quickMotion')==0)
        & (pl.col('onFieldPosition').is_in(['hb','f']))
    )
    .then('pistol')
    .otherwise(None)
    .max().over('uniquePlayId')
    .fill_null('shotgun')
    .alias('qbAlignment'),
    
    pl.when(
        (pl.col('underCenter')==1)
        & (pl.col('onFieldPosition')=='q')
        & (pl.col('position')!='QB')
    )
    .then(1)
    .otherwise(0)
    .max().over('uniquePlayId')
    .alias('skillPlayerSneak'),
    
    pl.when(
        (pl.col('underCenter')==0)
        & (pl.col('onFieldPosition')=='q')
        & (pl.col('position')!='QB')
    )
    .then(1)
    .otherwise(0)
    .max().over('uniquePlayId')
    .alias('wildcat'),
])
```

```python
# convert non-traditional positions from hbs to fullbacks if they're much closer to the ball than the deepest back

minimumFullbackDistance=1.5
df = df.with_columns([
    pl.when(
        (pl.col('underCenter')==1)
        & (pl.col('quickMotion')==0)
        & (pl.col('onFieldPosition')=='hb')
        & (pl.col('yFromBall') > 
                   (pl.when(pl.col('onFieldPosition')=='hb').then(pl.col('yFromBall')).otherwise(0)
                   .min().over('uniquePlayId')) + minimumFullbackDistance)      
    )
    .then('f')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition')
])
```

```python
# convert TEs close behind the QB in QB sneak scenarios to fullbacks, like in 2022100911-1200

df = df.with_columns([
    pl.when(
        (pl.col('underCenter')==1)
        & (pl.col('behindCenter')==1)
        & (pl.col('quickMotion')==0)
        & (pl.col('onFieldPosition').is_in(['ry','ly']))    
    )
    .then('f')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition')
])
```

```python
offsetFBThreshold = .5
df = df.with_columns(
    pl.when(
        (pl.col('xFromBall') >= 0)
        & (pl.col('onFieldPosition') == 'f')
        & (abs(pl.col('xFromBall') - pl.col('cX')) > offsetFBThreshold)
    )
    .then('rf')
    .when(
        (pl.col('xFromBall') < 0)
        & (pl.col('onFieldPosition') == 'f')
        & (abs(pl.col('xFromBall') - pl.col('cX')) > offsetFBThreshold)
    )
    .then('lf')
    .when(
        (pl.col('xFromBall') >= 0)
        & (pl.col('onFieldPosition') == 'hb')
        & (pl.col('behindCenter') == 0)
        & (pl.col('quickMotion') == 0)
    )
    .then('rhb')
    .when(
        (pl.col('xFromBall') < 0)
        & (pl.col('onFieldPosition') == 'hb')
        & (pl.col('behindCenter') == 0)
        & (pl.col('quickMotion') == 0)
    )
    .then('lhb')
    .when(
        (pl.col('onFieldPosition') == 'hb')
        & (pl.col('quickMotion') == 1)
    )
    .then('mhb')
    .otherwise(pl.col('onFieldPosition'))
    .alias('onFieldPosition') 
)
```

```python
# nominal personnel and functional personnel
df = df.with_columns([
    
    pl.when(pl.col('position').is_in(['RB','FB']))
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('nominalBacksCount'),
    
    pl.when(pl.col('onFieldPosition').is_in(['hb','rhb','lhb','mhb','f','lf','rf']))
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('functionalBacksCount'),
    
    pl.when(pl.col('position')=='TE')
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('nominalTEsCount'),
    
    pl.when(pl.col('onFieldPosition')=='ry')
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('rightFunctionalTEsCount'),
    
    pl.when(pl.col('onFieldPosition')=='ly')
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('leftFunctionalTEsCount'),
    
    pl.when(pl.col('position')=='TE')
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('nominalReceiversCount'),
    
    pl.when(
        (pl.col('onFieldPosition')=='wr')
        & (pl.col('xFromBall')>=0))
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('rightFunctionalReceiversCount'),
    
    pl.when(
        (pl.col('onFieldPosition')=='wr')
        & (pl.col('xFromBall')<0))
    .then(1)
    .otherwise(0)
    .sum().over('uniquePlayId')
    .alias('leftFunctionalReceiversCount'),
])


df = df.with_columns([
    (pl.col('rightFunctionalTEsCount')
     + pl.col('leftFunctionalTEsCount')).alias('functionalTEsCount'),
    
    (pl.col('rightFunctionalReceiversCount')
     + pl.col('leftFunctionalReceiversCount')).alias('functionalReceiversCount')
])

df = df.with_columns([
    (pl.col('nominalBacksCount').cast(str)
     + pl.col('nominalTEsCount').cast(str)).alias('nominalPersonnel'),
    
    (pl.col('functionalBacksCount').cast(str)
     + pl.col('functionalTEsCount').cast(str)).alias('functionalPersonnel')
])
```

```python
# Run strength rules are
# 1-To TE advantage
# 2-To FB advantage
# 3-Opposite RB
# 4-Most number of WR
# 5-Default to R

df = df.with_columns([
    # 1-To TE advantage
    pl.when(pl.col('rightFunctionalTEsCount') > pl.col('leftFunctionalTEsCount'))
    .then('r')
    .when(pl.col('rightFunctionalTEsCount') < pl.col('leftFunctionalTEsCount'))
    .then('l')
    
    
    # 2-To FB advantage
    .when(pl.col('rfCount') > pl.col('lfCount'))
    .then('r')
    .when(pl.col('rfCount') < pl.col('lfCount'))
    .then('l')
    
    
    # 3-Opposite RB advantage
    .when(pl.col('rhbCount') < pl.col('lhbCount'))
    .then('r')
    .when(pl.col('rhbCount') > pl.col('lhbCount'))
    .then('l')
    
    
    # 4-Most number of WR
    .when(pl.col('rightFunctionalReceiversCount') > pl.col('leftFunctionalReceiversCount'))
    .then('r')
    .when(pl.col('rightFunctionalReceiversCount') < pl.col('leftFunctionalReceiversCount'))
    .then('l')
    
    # 5-Default to R
    .otherwise('r')
    
    .alias('runStrength')
])
```

```python
def calculate_te_pulls(df, position):
    
    column_name = position +'Pull'
    
    
    df = df.with_columns([
        pl.lit(None).alias(column_name)
    ])
    
    if position.startswith('l'):
        
        if position=='ly1':
            one_right='lt'
        elif position=='ly2':
            one_right='ly1'
        elif position=='ly3':
            one_right='ly2'
          
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1.5) &
                (pl.col(position+'AdjustedY')<pl.col('ltAdjustedY')) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')) &
                (pl.col(position+'RelativeX')>1) 
                
            ).then('R')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
        
        # check for escort motion
        df = df.with_columns([
            pl.when(
                (pl.col(position+'QuickMotion')==1) &
                ((pl.col(position+'AdjustedO')<=75)  |
                (pl.col(position+'AdjustedO')>=270))
            ).then('ER') #Escort Right
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
        
        # check for counter bluff + verify orientation and relative position of non-bluffs
        df = df.with_columns([

            # counter bluff check
            pl.when(
                (pl.col(column_name).is_in(['R','ER'])) &
                (pl.col(position+'BluffCheckX')<pl.col(position+'AdjustedX')-4)
            ).then('BL')

            # verify relative position and orientation of non-bluffs to consider it a true pull
            .when(
                (pl.col(column_name)=='R') &

                ((pl.col(position+'AdjustedO')<=75)  |
                (pl.col(position+'AdjustedO')>=270)) &

                ((pl.col(position+'AdjustedX')>pl.col('ltAdjustedX')-1) |
                (pl.col(position+'AdjustedX')>pl.col(one_right+'AdjustedX')-1))

            ).then('R')
            
            .when(pl.col(column_name)=='ER').then('ER') #retain escort pulls
            .otherwise(pl.lit(None)).alias(column_name)
        ])

        
    elif position.startswith('r'):
        
        if position=='ry1':
            one_left='rt'
        elif position=='ry2':
            one_left='ry1'
        elif position=='ry3':
            one_left='ry2'
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1.5) &
                (pl.col(position+'AdjustedY')<pl.col('rtAdjustedY')) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')) &
                (pl.col(position+'RelativeX')<-1) 
            ).then('L')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
        
        # check for escort motion
        df = df.with_columns([
            pl.when(
                (pl.col(position+'QuickMotion')==1) & 
                (pl.col(position+'AdjustedO')>=105)  &
                (pl.col(position+'AdjustedO')<=270)
            ).then('EL') #Escort Left
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
        
        # check for counter bluff + verify orientation and relative position of non-bluffs
        df = df.with_columns([

            # counter bluff check
            pl.when(
                (pl.col(column_name).is_in(['L','EL'])) &
                (pl.col(position+'BluffCheckX')>pl.col(position+'AdjustedX')+4) 
            ).then('BR') #Bluff Right


            # verify relative position and orientation of non-bluffs to consider it a true pull
            .when(
                (pl.col(column_name)=='L') &

                (pl.col(position+'AdjustedO')>=105)  &
                (pl.col(position+'AdjustedO')<=270) &

                ((pl.col(position+'AdjustedX')<pl.col('rtAdjustedX')+1) |
                (pl.col(position+'AdjustedX')<pl.col(one_left+'AdjustedX')+1))

            ).then('L')
            
            .when(pl.col(column_name)=='EL').then('EL') #retain escort pulls
            .otherwise(pl.lit(None)).alias(column_name)
        ])


        
        
        
    return df
```

```python
def calculate_adjacent_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<=1) &
                (pl.col(position+'RelativeY')-.5 <= pl.col(one_right+'RelativeY')) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(position+'S')>2) &
                (pl.col(one_right+'Pull')=='R') &
                ((pl.col(position+'AdjustedO')<=75)  |
                (pl.col(position+'AdjustedO')>=270)) &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_right+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_right+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('R')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<1) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(position+'RelativeY')-.5 <= pl.col(one_left+'RelativeY')) &
                (pl.col(position+'S')>2) &
                (pl.col(one_left+'Pull')=='L') &
                (pl.col(position+'AdjustedO')>=105)  &
                (pl.col(position+'AdjustedO')<=270)  &
                (
                    ((pl.col(position+'AdjustedX')-pl.col(one_left+'AdjustedX')).pow(2) +
                    (pl.col(position+'AdjustedY')-pl.col(one_left+'AdjustedY')).pow(2)).sqrt() < 2.25
                ) # pythagoreon
            ).then('L')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```

```python
def calculate_pulls(df, position):
    linemen = list(linemen_positions.values())
    position_index = linemen.index(position)
    
    column_name = position +'Pull'
    
    df = df.with_columns([
        pl.lit(None).alias(column_name)
    ])
    
    if position != 'rt':
        one_right=linemen[position_index+1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')+.1) &
                (pl.col(position+'RelativeX')>1) &
                (pl.col(one_right+'RelativeX')<1) &
                (pl.col(position+'AdjustedX')>pl.col(one_right+'AdjustedX')-.5) &
                (pl.col(position+'S')>1.75) &
                ((pl.col(position+'AdjustedO')<=90)  |
                (pl.col(position+'AdjustedO')>=270))
            ).then('R')
            
            # crunch check
            .when(
                (pl.col(position+'Position').is_in(['lg','rg'])) &
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')) &
                (pl.col(one_right+'RelativeX')<1.5) &
                ((pl.col(position+'AdjustedO')<=60)  |
                (pl.col(position+'AdjustedO')>=315))
            ).then('CR')
            
            # tier 2 crunch check
            .when(
                (pl.col(position+'Position').is_in(['lg','rg'])) &
                (pl.col(position+'RelativeY')<1) &
                (pl.col(position+'AdjustedY')<pl.col(one_right+'AdjustedY')) &
                (pl.col(one_right+'RelativeX')<1.5) &
                ((pl.col(position+'AdjustedO')<=90)  |
                (pl.col(position+'AdjustedO')>=315))
            ).then('CR2')
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
        
    if position != 'lt':
        one_left=linemen[position_index-1]
        
        df = df.with_columns([
            pl.when(
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')+.1) &
                (pl.col(position+'RelativeX')<-1) &
                (pl.col(one_left+'RelativeX')>-1) &
                (pl.col(position+'AdjustedX')<pl.col(one_left+'AdjustedX')+.5) &
                (pl.col(position+'S')>1.75) &
                (pl.col(position+'AdjustedO')>=90)  &
                (pl.col(position+'AdjustedO')<=270)
            ).then('L')
            
            # crunch check
            .when(
                (pl.col(position+'Position').is_in(['lg','rg'])) &
                (pl.col(position+'RelativeY')<.5) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')) &
                (pl.col(one_left+'RelativeX')>-1.5) &
                (pl.col(position+'AdjustedO')>=120)  &
                (pl.col(position+'AdjustedO')<=225)
            ).then('CL')
            
            # tier 2 crunch check
            .when(
                (pl.col(position+'Position').is_in(['lg','rg'])) &
                (pl.col(position+'RelativeY')<1) &
                (pl.col(position+'AdjustedY')<pl.col(one_left+'AdjustedY')) &
                (pl.col(one_left+'RelativeX')>-1.5) &
                (pl.col(position+'AdjustedO')>=90)  &
                (pl.col(position+'AdjustedO')<=225)
            ).then('CL2')
            
            .otherwise(pl.col(column_name)).alias(column_name)
        ])
    
    return df
```

```python
crunchRight = (yPullLeft &
    (pl.col('rgPull').is_in(['L','CL'])) &
    (pl.col('lgPull').is_in(['L','CL'])) &
    undiagnosed)


crunchLeft = (yPullRight &
    (pl.col('rgPull').is_in(['R','CR'])) &
    (pl.col('lgPull').is_in(['R','CR'])) &
    undiagnosed)


# allow CR2s and CL2s, but wham must hit
crunchRightTier2 = (whamRight & 
    (pl.col('rgPull').is_in(['L','CL','CL2'])) &
    (pl.col('lgPull').is_in(['L','CL','CL2'])) &
    undiagnosed)


crunchLeftTier2 = (whamLeft & 
    (pl.col('rgPull').is_in(['R','CR','CR2'])) &
    (pl.col('lgPull').is_in(['R','CR','CR2'])) &
    undiagnosed)


# allow CR2s and CL2s, but trap must hit
crunchRightTier3 = (yPullRight & trapLeft & 
    (pl.col('lgPull').is_in(['L','CL','CL2'])) &
    undiagnosed)


crunchLeftTier3 = (yPullRight & trapRight &
    (pl.col('rgPull').is_in(['R','CR','CR2'])) &
    undiagnosed)


crunchRight = crunchRight | crunchRightTier2 | crunchRightTier3
crunchLeft = crunchLeft | crunchLeftTier2 | crunchLeftTier3
```

```python
fCrunchRight = ((pl.col('rfCount')>=1) &
    (pl.col('rgPull').is_in(['L','CL','CL2'])) &
    (pl.col('lgPull').is_in(['L','CL','CL2'])) &
    (pl.col('directionalTechnique').is_in(['4ir','3r','2r'])) &
    first_blocked_by('rf') &
    undiagnosed)


fCrunchLeft = ((pl.col('lfCount')>=1) &
    (pl.col('rgPull').is_in(['R','CR','CR2'])) &
    (pl.col('lgPull').is_in(['R','CR','CR2'])) &
    (pl.col('directionalTechnique').is_in(['4il','3l','2l'])) &
    first_blocked_by('lf') &
    undiagnosed)
```

```python
trapRight = trapRight & ~crunchLeft & (pl.coalesce(pl.col('ltPull'),pl.lit('none'))!='R')
trapLeft = trapLeft & ~crunchRight & (pl.coalesce(pl.col('rtPull'),pl.lit('none'))!='L')
```

```python
defense = defense.with_columns([
    pl.when(crunchRight | crunchLeft)
    .then('crunch')
    .when(fCrunchRight | fCrunchLeft)
    .then('fCrunch')
    
    .otherwise(pl.col('play'))
    .alias('play'),
    
    
    pl.when(crunchRight | fCrunchRight).then('r')
    .when(crunchLeft | fCrunchLeft).then('l')
    
    .otherwise(pl.col('playside'))
    .alias('playside'),
])

defense = save_plays(defense)
```

```python
spillBoxDentDict = {
    'r headup': 'dent',
    'l headup': 'dent',
    'r right': 'box',
    'l left': 'box',
    'r left': 'spill',
    'l right': 'spill',
}

results = results.select('uniquePlayId','nflId','nflIdDefender','blockStartFrame','playside')\
.join(blocks,
     on=['uniquePlayId','nflId','nflIdDefender']
     )\
.filter(pl.col('framesSinceSnap').is_between(pl.col('blockStartFrame'), pl.col('blockStartFrame')+2))\
.select('uniquePlayId','gameId','playId','nflId','nflIdDefender','blockStartFrame','framesSinceSnap',
        'playside',
        (pl.col('playside') + pl.lit(' ') 
         + pl.col('blockLocation')).map_dict(spillBoxDentDict).alias('defenseType')
       )\
.with_columns([
    pl.when(pl.col('defenseType')=='box').then(1).otherwise(0).max().over('uniquePlayId').alias('boxIndicator'),
    pl.when(pl.col('defenseType')=='dent').then(1).otherwise(0).max().over('uniquePlayId').alias('dentIndicator'),
])\
.filter(pl.col('framesSinceSnap')==pl.col('blockStartFrame')+2)\
.with_columns([
    pl.when(
        (pl.col('defenseType')=='spill') & (pl.col('boxIndicator')==1)
    ).then('box')
    .when(
        (pl.col('defenseType')=='box') & (pl.col('dentIndicator')==1)
    ).then('dent')
    .otherwise(pl.col('defenseType'))
    .alias('defenseType'),
    
    pl.col('defenseType').alias('oldDefenseType')
])

results
```
