```python
tracking = pl.read_csv('nfl-big-data-bowl-2024/tracking_week*.csv',infer_schema_length=10000)
tracking = tracking.with_columns(
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str)).alias('uniquePlayId'),
    (pl.col('gameId').cast(str) + '-'
     + pl.col('playId').cast(str) + '-'
     + pl.col('nflId').cast(str)).alias('uniquePlayerId'),
)
# tracking=tracking.filter(pl.col('club')!='football')
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('playDirection')=='right').then(53.3-pl.col('y')).otherwise(pl.col('y')).alias('adjustedX'),
    pl.when(pl.col('playDirection')=='right').then(pl.col('x')).otherwise(120-pl.col('x')).alias('adjustedY'),
#     pl.when(pl.col('playDirection')=='right').then(pl.col('dir').cast(pl.Float64)).otherwise(180-pl.col('dir').cast(pl.Float64)).alias('adjustedDir'),
#     pl.when(pl.col('playDirection')=='right').then(pl.col('o').cast(pl.Float64)).otherwise(180-pl.col('o').cast(pl.Float64)).alias('adjustedO'),
])

# tracking=tracking.with_columns([
#     pl.when(pl.col('adjustedDir')<0).then(360+pl.col('adjustedDir')).otherwise(pl.col('adjustedDir')).alias('adjustedDir'),
#     pl.when(pl.col('adjustedO')<0).then(360+pl.col('adjustedO')).otherwise(pl.col('adjustedO')).alias('adjustedO'),
# ])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('event')=='ball_snap').then(pl.col('frameId')).otherwise(-1).alias('startingFrameId'),
])
tracking=tracking.with_columns([
    pl.col('startingFrameId').max().over(pl.col('uniquePlayId')),
])
tracking=tracking.with_columns([
    (pl.col('frameId') - pl.col('startingFrameId')).alias('framesSinceSnap'),
])
tracking
```

```python
tracking=tracking.with_columns([
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedX')).otherwise(-1000).alias('startingX'),
    pl.when(pl.col('framesSinceSnap')==0).then(pl.col('adjustedY')).otherwise(-1000).alias('startingY')
])
tracking=tracking.with_columns([
    pl.col('startingX').max().over(pl.col('uniquePlayerId')),
    pl.col('startingY').max().over(pl.col('uniquePlayerId'))
])
tracking=tracking.with_columns([
    (pl.col('adjustedX') - pl.col('startingX')).alias('relativeX'),
    (pl.col('adjustedY') - pl.col('startingY')).alias('relativeY'),
])
# tracking=tracking.with_columns([
#     pl.when(pl.col('club')=='football').then(pl.col('startingX')).otherwise(-1000).alias('footballStartingX')
# ])
# tracking=tracking.with_columns([
#     pl.col('footballStartingX').max().over(pl.col('uniquePlayId'))
# ])
tracking
```

```python
def convert_frame(frame):
    components =a frame.split(':')
    return int(components[0]) * 10 + int(components[1]) // 3
```

```python
def label_blocking(df,gameId,playId,frameInfo):
    return df.with_columns([
        pl.when(
            (pl.col('gameId') == gameId) &
            (pl.col('playId') == playId) &
            (pl.col('jerseyNumber') == frameInfo['blockerNo']) &
            (pl.col('jerseyNumberDefender') == frameInfo['defenderNo']) &
            (pl.col('framesSinceSnap').is_between(frameInfo['startFrame'], frameInfo['endFrame'])) 
        )
        .then(frameInfo['blockType'])
        .otherwise(pl.col('blockType'))
#         .otherwise(pl.lit(0))
        .alias('blockType')
    ])
```
